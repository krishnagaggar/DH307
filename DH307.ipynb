```python
df = pd.read_csv(r'C:\Users\kgagg\Desktop\krishna\semester6\dh307\Dataset 2.csv')
display(df)
mask = df == 1.000000
selected_columns = df.loc[:, mask.any()]
df_1 = selected_columns
df_1 = df_1[1:]
mask = df == 2.000000
selected_columns = df.loc[:, mask.any()]
df_2 = selected_columns
df_2 = df_2[1:]
df_1.loc[:, 'Sample'] = df['Sample'][1:]
df_2.loc[:, 'Sample'] = df['Sample'][1:]
df_1.set_index('Sample', inplace=True)
df_2.set_index('Sample', inplace=True)
```

```python
display(df_2.describe())
```

```python
plt.figure(figsize=(25, 15))

for column in df_2.columns:
    plt.subplot(3, 3, df_2.columns.get_loc(column) + 1)  # Adjust the subplot layout as needed
    sns.kdeplot(df_2[column], color='skyblue', fill=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
df_2_transformed = df_2.apply(lambda x: np.log(x) if np.issubdtype(x.dtype, np.number) else x)
```

```python
df_2_normalized = df_2_transformed.copy()  

for column in df_2.columns:
    if np.issubdtype(df_2[column].dtype, np.number):   
        median_value = df_2[column].median()
        df_2_normalized[column] = df_2[column] - median_value

df_2_normalized.describe()
```

```python
plt.figure(figsize=(25, 15))

for column in df_2_normalized.columns:
    plt.subplot(3, 3, df_2_normalized.columns.get_loc(column) + 1)  # Adjust the subplot layout as needed
    sns.kdeplot(df_2_normalized[column], color='skyblue', fill=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

```python
correlation_matrix = df_2_normalized.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title("Correlation Heatmap")
plt.show()
```

```python
def median_mad_scaling(column_data):
    median_value = np.median(column_data)
    mad_value = np.median(np.abs(column_data - median_value))
    
    scaled_data = (column_data - median_value) / mad_value
    
    return scaled_data

# Apply Median-MAD scaling to each column
df_scaled = df_2_normalized.apply(median_mad_scaling)

# Display the first few rows of the scaled DataFrame
display(df_scaled.describe())
```

```python
plt.figure(figsize=(25, 15))

for column in df_scaled.columns:
    plt.subplot(3, 3, df_scaled.columns.get_loc(column) + 1)  # Adjust the subplot layout as needed
    sns.kdeplot(df_scaled[column], color='skyblue', fill=True)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
